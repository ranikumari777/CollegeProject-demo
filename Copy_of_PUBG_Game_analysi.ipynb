Steps for Analysis in Python Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```



To perform a hands-on analysis of PlayerUnknown’s Battlegrounds (PUBG) game data, we need a dataset that provides information about players, their performances, and match outcomes. Using Python, we'll explore player statistics and draw insights to understand factors that may influence player success in the game.

Steps for Analysis in Python Step 1: Import Libraries

Step 2: Load the Dataset In a PUBG game,players get ranked at the end of the game (winPlacePerc) based on how many other teams are still alive when they are eliminated. In game, players can pick up different munitions, revive downed-but-not-out (knocked) teammates, drive vehicles, swim, walk, shoot, and experience all of the consequences -- such as falling too far or running themselves over and eliminating themselves.


```python
df = {
    'kills': [3, 0, 5, 1, 2],
    'damageDealt': [450.0, 100.0, 950.0, 300.0, 650.0],
    'winPlacePerc': [0.76, 0.22, 0.89, 0.45, 0.60],
    'boosts': [4, 0, 2, 1, 3],
    'heals': [3, 1, 5, 2, 3],
    'walkDistance': [1200, 200, 1500, 800, 1100],
    'rideDistance': [0, 300, 0, 1000, 200],
    'swimDistance': [0, 0, 0, 50, 20],
    'headshotKills': [1, 0, 2, 0, 1],
    'killStreaks': [1, 0, 2, 0, 1]
}

# Convert dictionary to DataFrame
data = pd.DataFrame(df)
```

Step 3: Explore the Data

```python
data.head()
```

```python
print(data.info())
print(data.describe())
```

```python
print(data.isnull().sum())
```



If there are missing values, you can decide to fill them with the mean, median, or drop those rows.

Step 5: Data Visualization and Analysis Let's analyze some factors that might influence the win placement percentage (winPlacePerc).

    Relationship between Kills and Win Placement



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='kills', y='winPlacePerc', data=data)
plt.title('Relationship between Kills and Win Placement')
plt.xlabel('Kills')
plt.ylabel('Win Placement Percentage')
plt.show()
```



Explanation: This scatter plot helps us see if players with higher kills tend to have a better win placement percentage.

    Effect of Damage Dealt on Win Placement



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='damageDealt', y='winPlacePerc', data=data)
plt.title('Effect of Damage Dealt on Win Placement')
plt.xlabel('Damage Dealt')
plt.ylabel('Win Placement Percentage')
plt.show()
```



Explanation: This plot shows if there’s a correlation between the amount of damage a player dealt and their placement in the match.

    Boosts and Heals Usage: Using boosts and heals can be essential for survival. Let's see how these impact win placement.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='boosts', y='winPlacePerc', data=data, label='Boosts')
sns.scatterplot(x='heals', y='winPlacePerc', data=data, label='Heals', color='orange')
plt.title('Effect of Boosts and Heals on Win Placement')
plt.xlabel('Boosts / Heals')
plt.ylabel('Win Placement Percentage')
plt.legend()
plt.show()
```



Explanation: This visualization compares the usage of boosts and heals with win placement, indicating whether these items contribute to better placement in the game.

    Movement and Win Placement We’ll analyze how the distances traveled (walk, ride, swim) impact the player’s success.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='walkDistance', y='winPlacePerc', data=data, label='Walk Distance')
sns.scatterplot(x='rideDistance', y='winPlacePerc', data=data, label='Ride Distance', color='green')
sns.scatterplot(x='swimDistance', y='winPlacePerc', data=data, label='Swim Distance', color='purple')
plt.title('Effect of Movement on Win Placement')
plt.xlabel('Distance')
plt.ylabel('Win Placement Percentage')
plt.legend()
plt.show()
```



Explanation: This chart shows if players who travel more on foot, by vehicle, or by swimming are more likely to place well.

Step 6: Analyze Headshot Kills and Kill Streaks Let’s analyze if getting headshots or consecutive kills (kill streaks) has an impact on win placement.


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='headshotKills', y='winPlacePerc', data=data)
plt.title('Effect of Headshot Kills on Win Placement')
plt.xlabel('Headshot Kills')
plt.ylabel('Win Placement Percentage')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='killStreaks', y='winPlacePerc', data=data)
plt.title('Effect of Kill Streaks on Win Placement')
plt.xlabel('Kill Streaks')
plt.ylabel('Win Placement Percentage')
plt.show()
```

Explanation: These box plots show the distribution of win placement for different levels of headshot kills and kill streaks. High headshot kills or kill streaks may contribute to a better ranking.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = {
    'kills': [3, 0, 5, 1, 2],
    'damageDealt': [450.0, 100.0, 950.0, 300.0, 650.0],
    'winPlacePerc': [0.76, 0.22, 0.89, 0.45, 0.60],
    'boosts': [4, 0, 2, 1, 3],
    'heals': [3, 1, 5, 2, 3],
    'walkDistance': [1200, 200, 1500, 800, 1100],
    'rideDistance': [0, 300, 0, 1000, 200],
    'swimDistance': [0, 0, 0, 50, 20],
    'headshotKills': [1, 0, 2, 0, 1],
    'killStreaks': [1, 0, 2, 0, 1]
}

# Convert dictionary to DataFrame
data = pd.DataFrame(df)

# Data exploration
data.head()
print(data.info())
print(data.describe())
print(data.isnull().sum())

# Visualization 1: Kills vs Win Placement
plt.figure(figsize=(10, 6))
sns.scatterplot(x='kills', y='winPlacePerc', data=data)
plt.title('Relationship between Kills and Win Placement')
plt.xlabel('Kills')
plt.ylabel('Win Placement Percentage')
plt.show()

# Visualization 2: Damage Dealt vs Win Placement
plt.figure(figsize=(10, 6))
sns.scatterplot(x='damageDealt', y='winPlacePerc', data=data)
plt.title('Effect of Damage Dealt on Win Placement')
plt.xlabel('Damage Dealt')
plt.ylabel('Win Placement Percentage')
plt.show()

# Visualization 3: Boosts and Heals vs Win Placement
plt.figure(figsize=(10, 6))
sns.scatterplot(x='boosts', y='winPlacePerc', data=data, label='Boosts')
sns.scatterplot(x='heals', y='winPlacePerc', data=data, label='Heals', color='orange')
plt.title('Effect of Boosts and Heals on Win Placement')
plt.xlabel('Boosts / Heals')
plt.ylabel('Win Placement Percentage')
plt.legend()
plt.show()

# Visualization 4: Movement (Walk, Ride, Swim Distance) vs Win Placement
plt.figure(figsize=(10, 6))
sns.scatterplot(x='walkDistance', y='winPlacePerc', data=data, label='Walk Distance')
sns.scatterplot(x='rideDistance', y='winPlacePerc', data=data, label='Ride Distance', color='green')
sns.scatterplot(x='swimDistance', y='winPlacePerc', data=data, label='Swim Distance', color='purple')
plt.title('Effect of Movement on Win Placement')
plt.xlabel('Distance')
plt.ylabel('Win Placement Percentage')
plt.legend()
plt.show()

# Visualization 5: Headshot Kills vs Win Placement
plt.figure(figsize=(10, 6))
sns.boxplot(x='headshotKills', y='winPlacePerc', data=data)
plt.title('Effect of Headshot Kills on Win Placement')
plt.xlabel('Headshot Kills')
plt.ylabel('Win Placement Percentage')
plt.show()

# Visualization 6: Kill Streaks vs Win Placement
plt.figure(figsize=(10, 6))
sns.boxplot(x='killStreaks', y='winPlacePerc', data=data)
plt.title('Effect of Kill Streaks on Win Placement')
plt.xlabel('Kill Streaks')
plt.ylabel('Win Placement Percentage')
plt.show()
```



Explanation Summary This code performs exploratory data analysis (EDA) on PUBG player statistics:

Kills and Damage Dealt: Higher kills and more damage dealt may improve win placement. Boosts and Heals: Using boosts and heals might help players survive longer. Movement Distances: More distance traveled (walking, riding, or swimming) may correlate with better placement. Headshots and Kill Streaks: High headshot count or achieving kill streaks could relate to higher placements.
